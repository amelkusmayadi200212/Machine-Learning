# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Memuat dataset
dataset = pd.read_csv('/content/drive/MyDrive/Dataset/Salary_Data.csv')

# Menampilkan 5 baris pertama dataset
print(dataset.head())

# Memisahkan fitur (YearsExperience) dan target (Salary)
X = dataset.iloc[:, :-1].values  # Kolom "YearsExperience"
Y = dataset.iloc[:, -1].values  # Kolom "Salary"
```

```python
# Memisahkan data menjadi data latih dan data uji (80% latih, 20% uji)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Membuat model regresi linear
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# Memprediksi hasil pada data uji
Y_pred = regressor.predict(X_test)
```

```python
# Menghitung Mean Squared Error dan R-squared
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# Visualisasi hasil pada data latih
plt.scatter(X_train, Y_train, color='red', label='Data Aktual')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Regresi Linear')
plt.title('Gaji vs Pengalaman (Data Latih)')
plt.xlabel('Tahun Pengalaman')
plt.ylabel('Gaji')
plt.legend()
plt.show()
```

```python
# Visualisasi hasil pada data uji
plt.scatter(X_test, Y_test, color='red', label='Data Aktual')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Regresi Linear')  # Garis dari model yang sama
plt.title('Gaji vs Pengalaman (Data Uji)')
plt.xlabel('Tahun Pengalaman')
plt.ylabel('Gaji')
plt.legend()
plt.show()
```
