# Polynomial Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Memuat dataset
dataset = pd.read_csv('/content/drive/MyDrive/Dataset/Position_Salaries.csv')

# Memisahkan fitur (Level) dan target (Salary)
X = dataset.iloc[:, 1:2].values  # Kolom "Level" sebagai fitur
Y = dataset.iloc[:, 2].values  # Kolom "Salary" sebagai target

print(dataset.head())
```

```python
# Membuat model regresi linear
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
# Mengubah fitur menjadi polinomial dengan degree 4
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)

# Membuat model regresi linier pada fitur polinomial
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
# Visualisasi hasil regresi linear
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualisasi hasil regresi polinomial
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Polynomial Regression (Degree 4)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Memprediksi gaji dengan model Linear Regression
lin_pred = lin_reg.predict([[6.5]])
print(f"Gaji Prediksi dengan Linear Regression: {lin_pred[0]}")

# Memprediksi gaji dengan model Polynomial Regression
poly_pred = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print(f"Gaji Prediksi dengan Polynomial Regression: {poly_pred[0]}")
```
