Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
# Mengatur gaya grafik seaborn untuk estetika lebih baik
sns.set(style="whitegrid")

def plot_pie_chart(column_name):

    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values


    plt.figure(figsize=(7,7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}')
    plt.show()

plot_pie_chart('gender')


plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
# Fungsi untuk menentukan apakah siswa lulus atau tidak berdasarkan skor
def is_pass(score):
  if score >= 75:
    return 'Pass'
  else:
    return 'Fail'

# Membuat kolom baru untuk status lulus/tidak lulus pada masing-masing pelajaran
df['math_status'] = df['math score'].apply(is_pass)
df['reading_status'] = df['reading score'].apply(is_pass)
df['writing_status'] = df['writing score'].apply(is_pass)

# Menghitung jumlah siswa yang lulus dan tidak lulus untuk setiap pelajaran
math_status_counts = df['math_status'].value_counts()
reading_status_counts = df['reading_status'].value_counts()
writing_status_counts = df['writing_status'].value_counts()

# Menampilkan hasilnya
print("Jumlah siswa yang lulus/tidak lulus untuk pelajaran Math:")
print(math_status_counts)
print("\nJumlah siswa yang lulus/tidak lulus untuk pelajaran Reading:")
print(reading_status_counts)
print("\nJumlah siswa yang lulus/tidak lulus untuk pelajaran Writing:")
print(writing_status_counts)
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
# Membuat kolom baru yang berisi rata-rata skor dari ketiga mata pelajaran
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

# Melihat korelasi antara rata-rata skor dengan variabel kategorikal
correlation_gender = df.groupby('gender')['average_score'].mean()
print("Korelasi antara gender dan rata-rata skor:\n", correlation_gender)

correlation_race = df.groupby('race/ethnicity')['average_score'].mean()
print("\nKorelasi antara race/ethnicity dan rata-rata skor:\n", correlation_race)

correlation_parental = df.groupby('parental level of education')['average_score'].mean()
print("\nKorelasi antara parental level of education dan rata-rata skor:\n", correlation_parental)

correlation_lunch = df.groupby('lunch')['average_score'].mean()
print("\nKorelasi antara lunch dan rata-rata skor:\n", correlation_lunch)

correlation_preparation = df.groupby('test preparation course')['average_score'].mean()
print("\nKorelasi antara test preparation course dan rata-rata skor:\n", correlation_preparation)


# Visualisasi Korelasi (contoh dengan gender)
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='average_score', data=df)
plt.title('Korelasi antara Gender dan Rata-rata Skor')
plt.show()
````
